```python
import hdgim
import numpy as np
import matplotlib.pyplot as plt

DIMENSION = 10000
DNA_SEQUENCE_LENGTH = 1000
DNA_SUBSEQUENCE_LENGTH = 20
NUM_OF_TRUE = 50
NUM_OF_FALSE = 50  
BIT_PRECISION = 2
NOISE_PROBABILTY = 0.1
```

```python
model = hdgim.HDGIM(DIMENSION, DNA_SEQUENCE_LENGTH, DNA_SUBSEQUENCE_LENGTH, NUM_OF_TRUE, NUM_OF_FALSE, BIT_PRECISION, NOISE_PROBABILTY)
```

```python
#{'A': 0, 'T': 1, 'C': 2, 'G': 3}
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Plot the histograms in each subplot
axes[0, 0].hist(model.base_hypervectors[0].numpy(), bins=30, density=True, alpha=0.7, edgecolor='black')
axes[0, 0].set_title('Value of Base Hypervector A')
axes[0, 0].set_xlabel('Value')
axes[0, 0].set_ylabel('Density')

axes[0, 1].hist(model.base_hypervectors[1].numpy(), bins=30, density=True, alpha=0.7, edgecolor='black')
axes[0, 1].set_title('Value of Base Hypervector T')
axes[0, 1].set_xlabel('Value')
axes[0, 1].set_ylabel('Density')

axes[1, 0].hist(model.base_hypervectors[2].numpy(), bins=30, density=True, alpha=0.7, edgecolor='black')
axes[1, 0].set_title('Value of Base Hypervector C')
axes[1, 0].set_xlabel('Value')
axes[1, 0].set_ylabel('Density')

axes[1, 1].hist(model.base_hypervectors[3].numpy(), bins=30, density=True, alpha=0.7, edgecolor='black')
axes[1, 1].set_title('Value of Base Hypervector G')
axes[1, 1].set_xlabel('Value')
axes[1, 1].set_ylabel('Density')
```

```python
model.binding()

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.hist(model.encoded_hypervector.numpy(), bins=30, density=True, alpha=0.7, edgecolor='black')
ax.set_title('Value of Encoded Hypervector')
ax.set_xlabel('Value')
ax.set_ylabel('Density')

plt.show()
```

```python
model.quantize()

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.hist(model.quantized_hypervector.numpy(), bins=30, density=True, alpha=0.7, edgecolor='black')
ax.set_title('Value of Quantized Hypervector')
ax.set_xlabel('Value')
ax.set_ylabel('Density')

plt.show()
```

```python
model.adding_noise()

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.hist(model.quantized_hypervector_with_noise.numpy(), bins=30, density=True, alpha=0.7, edgecolor='black')
ax.set_title('Value of Quantized Hypervector with Noise')
ax.set_xlabel('Value')
ax.set_ylabel('Density')

plt.show()
```
